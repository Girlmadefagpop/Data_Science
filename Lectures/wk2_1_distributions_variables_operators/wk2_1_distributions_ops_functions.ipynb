## Distributions - real and simulated

In this course we are interested in 
- how are real data distributed?
- how does this compare to a simulated world where we know what is happening?

### Are men and women's scores different?
- What would the distributions look like if we *know* the scores are the same?
  - not _exactly_ the same, but
  - the same $+$ some variation 


# The essential logic of the course

- ### simulate scores from a made-up world (e.g. a world where we know the _real_ difference is zero)
- ### compare to the real world
- ### How similar are the worlds?
  - In a quantitatively precise way
<br>

### simulated _sampling_ always involves some random element
- e.g. sampling people at random from a larger group where the scores are the same
- the **sampled** groups will not typically have the same scores

### compare the simulated scores to the actual scores
- Are they different enough so we can say there is a real difference?
  - What is different enough? (we will see how to decide this)
  - The basic idea, however, is that actual scores would be **unlikely** in our no-difference world
- Or are they very similar to scores from the world we simulated
  - (the world where we are sure there is actually no difference)
  - i.e. the scores are a pretty **likely** outcome from a no-difference world

### Operations are done in order and previous values are **not** updated

````python
meal_cost = 10
vat = .2
tip = 0.15

total_bill = meal_cost + (meal_cost * vat) + (meal_cost * tip)
print("total bill was: ",total_bill)

# oh no!  meal_cost was higher than I thought

meal_cost = 20
````

### what is ```total_bill``` now?

````python
print(total_bill)
````

### what??!??

### What happened?

- ```total_bill``` was calculated when ```meal_cost = 10```
- ```total_bill``` does not change just because ```meal_cost``` changes
- To change ```total_bill``` you have to re-calculate it

````python
# try it
meal_cost = 10
vat = .2
tip = 0.15

total_bill = meal_cost + (meal_cost * vat) + (meal_cost * tip)
print("total bill was: ",total_bill)

# oh no!  meal_cost was higher than I thought

meal_cost = 20
total_bill = meal_cost + (meal_cost * vat) + (meal_cost * tip)
print("new total bill is",total_bill)
````

### Order of operations or _precedence_

- Add 1 + 3
- Multiply the result by 5

````python
my_result = 1 + 3 * 5
print(my_result)
````

## WHAT!!?!

## GOLDEN COMPUTER RULE: Multiply/divide first -- Add/subtract after
## If in doubt, use parentheses

- Python will do what is **inside** parentheses first
- Then it will do things outside parentheses

````python
my_result = (1+3) * 5
print(my_result)
````

Does 1+3 before multiplying the result by 5
- Parentheses can be nested

````python
my_result = (((1+3)*5) + 10) * 100
print(my_result)
````

- Add 1+3
- Multiply result by 5
- Add 10
- Multiply by 100
