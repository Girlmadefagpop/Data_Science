## Answer to exercise 1

```python
import numpy as np

# complet your code here:

# Write a function to simulate 5 families, each having 4 children
# Return the percentage of families having 3 boys as 'percent_3_boys'

def simulate_3_boys_in_4():  
    # Initialise an array of 5 empty values
    n_boys_each_family = np.full(5, np.nan) 

    # For family 1, simulate the sex for their 4 children, with girl labelled as 0, boy as 1
    family1 = np.random.randint(low=0, high= 2, size= 4)
    
    n_boys_each_family[0] = np.sum(family1) # Store the number of boys in position 0 of the array

    #Repeat for family 2-5
    family2 = np.random.randint(low=0, high= 2, size= 4)
    n_boys_each_family[1] = np.sum(family2) # Store the number of boys in position 1 of the array

    family3 = np.random.randint(low=0, high= 2, size= 4)
    n_boys_each_family[2] = np.sum(family3) # Store the number of boys in position 2 of the array

    family4 = np.random.randint(low=0, high= 2, size= 4)
    n_boys_each_family[3] = np.sum(family4) # # Store the number of boys in position 3 of the array

    family5 = np.random.randint(low=0, high= 2, size= 4)
    n_boys_each_family[4] = np.sum(family5) # Store the number of boys in position 4 of the array

    #Count how many 3s are there in the array and divide the count by 5
    percent_3_boys =  sum(n_boys_each_family ==3) /5
    
    return percent_3_boys
```

```python
# test your function here
# set the same seed so our results are comparable. If you remove this line, the result will change each time you run
np.random.seed(14102025) 
print (simulate_3_boys_in_4())
```

## Answer to exercise 2

```python
# complet your code here:

# Write a function to simulate 5 families, each having 'n_chi' children. 
# Return the percentage of families having 'n_boys' boys as 'percent_n_boys'

# We obtain this answer by replacing a few variables in the function above:
# size = 4 was replaced by size = n_chil
# sum(n_boys_each_family ==3) /5 was replaced by sum(n_boys_each_family == n_boys) /5


def simulate_boys(n_boys, n_chi):   
    # complet your code here:

    # Initialise an array of 5 empty values
    n_boys_each_family = np.full(5, np.nan) 

    # For family 1, simulate the sex for their 'n_chi' children, with girl labelled as 0, boy as 1
    family1 = np.random.randint(low=0, high= 2, size= n_chi)
    
    n_boys_each_family[0] = np.sum(family1)  # Store the number of boys in position 0 of the array

    #Repeat for family 2-5
    family2 = np.random.randint(low=0, high= 2, size= n_chi)
    n_boys_each_family[1] = np.sum(family2) 

    family3 = np.random.randint(low=0, high= 2, size= n_chi)
    n_boys_each_family[2] = np.sum(family3)

    family4 = np.random.randint(low=0, high= 2, size= n_chi)
    n_boys_each_family[3] = np.sum(family4)

    family5 = np.random.randint(low=0, high= 2, size= n_chi)
    n_boys_each_family[4] = np.sum(family5)

    #Count how many 'n_boys' are there in the array and divide the count by 5
    percent_n_boys =  sum(n_boys_each_family == n_boys) /5

    return  percent_n_boys
```

```python
# test your function here
# set the same seed so our results are comparable. 
# Remove the following line to see that your results will change each time you run
np.random.seed(14102025) 
print (simulate_boys(3, 4))
```
