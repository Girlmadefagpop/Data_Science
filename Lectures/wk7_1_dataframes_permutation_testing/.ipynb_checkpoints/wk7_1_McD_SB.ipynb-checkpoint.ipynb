```python
import pandas as pd # load the usual suspects
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('bmh')

# read in the menu data and check that it looks OK
sb_vs_mac = pd.read_csv("McD_vs_StarB_menus.csv")
sb_vs_mac.head()
```

```python
sb_vs_mac.tail()
```

```python
# Get only MickyD items - use a boolean expression
# use 'copy' because you are creating a new dataframe

McD_df = sb_vs_mac[sb_vs_mac["Restaurant"] == "McDonalds"].copy()
McD_df.head()
```

```python
# get only starbucks items - use a boolean expression
# use 'copy' because you are creating a new dataframe

SB_df = sb_vs_mac[sb_vs_mac["Restaurant"] == "Starbucks"].copy()
SB_df.head()
```

```python
# plot fat values
# note how you set the color of the bars
# 'alpha' sets how transparent the color should be
#        -50% means 50% of the other color shows through

plt.hist(SB_df["Fat"],color="green",alpha=0.5)
plt.hist(McD_df["Fat"],color="red",alpha=0.5)
```

### Improvements -- hard to compare, bins are different sizes, bars are very different heights
#### Why the height difference?  
- McDonalds has many more items (taller bins)
#### The bin size is different for Starbucks and McDonalds (smaller bins for SB)
- set one standard bin size for both histograms
- notice the range of values you need to cover on the x axis (0 - 120) 


```python
# generate a series of numbers between 0 and 120 - 
# set the step size to the bin size you want
# this will be the numbers that define our bins
# (experiment with larger or smaller bins)

my_bins = np.arange(0,120,5)
my_bins
```

```python
# improvements -- make bin sizes match for the two histograms

my_bins = np.arange(0,120,5)
# plot fat values

plt.hist(McD_df["Fat"],color="red",alpha=0.5,bins=my_bins)
plt.hist(SB_df["Fat"],color="green",alpha=0.5,bins=my_bins)
```

### Hard to see the relationship -- McDonalds has more counts so most bars are taller

We would like the total of all bar heights to be about the same for SM and MD so we can compare
- McDonalds bars are higher just because the McDonalds dataset is bigger

- Use the option `density = True` to make bars into a % of the total counts
    - (instead of the raw number of counts)
- `alpha` is the level of transparency (so we can see two colours on top of each other)

```python
plt.hist(McD_df["Fat"],color="red",alpha=0.5, density=True, bins=my_bins)
plt.hist(SB_df["Fat"],color="green",alpha=0.5, density=True, bins=my_bins)

# label the plot properly
plt.xlabel("Fat")
plt.ylabel("Frequency - % of total count")
plt.title("McDonalds (red) vs Starbucks (green)")
```

### Better -- now we can see what we need to see.  What do you conclude so far?
- What is shape of McDonalds' distribution?
- What is the shape of Starbuck's distribution?
- What do you take from the shapes?
