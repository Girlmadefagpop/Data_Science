### Simulate [num1] boys in families of size [num2]

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# get the number of boys in a family of size "size of 

def number_of_boys(size_family):
    # NOTE - size_family can only be seen inside the function
    # size_of_family (below) is outside the function
    # it is a separate variable
    family_members = np.full(size_family,np.nan) 
    for i in np.arange(0,size_family):
        family_members[i] = np.random.randint(low=0, high=2)
    num_boys = int(np.sum(family_members))
    return(num_boys)

size_of_family = 6
boys_in_current_family = number_of_boys(size_of_family)
boys_in_current_family
````

````python
def simulate_X_families(number_of_families,family_size):
    boys_in_families_result = np.full(number_of_families,np.nan)
    for i in np.arange(0,number_of_families):
        boys_in_families_result[i] = number_of_boys(family_size)
    return(boys_in_families_result)

boys_in_families = simulate_X_families(100,15)
print(boys_in_families)
# this calculates the number of boys in families 
# We simulate ```number_of_families``` families, 
# each of size ``family_size```
# We count the number of boys in each and store that
# We return an array that has the number of boys in each family
````

````python
### Let's see how variable the number of boys is
#### We can run our simulation of 100 families

# Henry VIII had at least 15 children - four survived
# and 11 were stillborn or died in infancy
# 2 surviving children were boys

# How likely is that?
number_of_families_to_simulate = 100
family_size = 15
boys_in_families = simulate_X_families(number_of_families_to_simulate,
                                       family_size)

actual_num_boys = 2 
plt.hist(boys_in_families)
plt.axvline(actual_num_boys,color="red")

# try with 100, then try with 1000 or 10000
````

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
def simulate_jury(percent_black):
    percent_white = 100-percent_black
    # one way to simulate 26% black - a different way is below
    jury_members = np.random.choice([1,0],size=12,p=[.26,.74])
    return(jury_members)
````

````python
simulate_jury(26)
````

````python
# counting black members

# note!  a different way to simulate
jury = np.random.randint(low=0, high=100, size=12)
print(jury)
black_members = jury < 26
print(black_members)
num_black = np.sum(black_members)
print(num_black)
````

### Solution

````python

# prepare for loop
percent_black = 26
number_of_simulations = 1000
num_black_jurors = np.full(number_of_simulations,np.nan)

# loop
for i in np.arange(0,number_of_simulations):
    one_jury = simulate_jury(percent_black)
    num_black_jurors[i] = np.sum(one_jury) # count black members

# show result - first plot then probability
plt.hist(num_black_jurors)
plt.axvline(0.5,color="red") # 0.5 to mark middle of the 0 - 1 space
````

````python
# probability

count_eq_zero = np.sum(num_black_jurors == 0)
print("number of juries with 0 black members out of ",number_of_simulations," juries = ",count_eq_zero)
p_fair = count_eq_zero/number_of_simulations
print("Probability that jury selection was fair: ",p_fair)
````
