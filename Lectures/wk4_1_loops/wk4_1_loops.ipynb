### The process - conceptually:
#### Write code to simulate an outcome we can define, e.g.
**Mosquitoes and beer**

  - If the actual result would rarely occur if the jury was fair, conclude that the jury was unfair  
<br>

## Quick topic - floating point and integer numbers

# Why are there two types of numbers?

Computers represent integers (whole numbers) and floating point numbers (decimals) differently.

#### Integer representation

A series of zeros and ones (8/16/32/64).
- first digit - $2^0$ = 1 if value is 1, or else 0  
- second digit - $2^1$ = 2 if value is 1 or else 0
- third digit - $2^2$ = 4  
- fourth digit - $2^3$ = 8
- etc.
- last digit is a **sign** digit: 1 = -1; 0 = +1

#### Floating point numbers - decimals

Decimals are actually quite compliated to represent and are a topic all on their own.  If you want to know more about the nitty-gritty, see here: <https://en.wikipedia.org/wiki/Floating-point_arithmetic>

Floating point representation is similar to what you may know as _scientific notation_.
- There is a mantissa (the part behind the decimal point)
- There is an exponent (what the base is raised to)

If the base is 10, a number like 34.567 is represented like this:

The computer version stores just the decimal (mantissa) and the exponent.  Here is an example 8-bit floating-point number:

The last line is an example 32-bit floating-point number

#### That is the why (maybe more than you wanted to know) -- So what?

- Sometimes you need to know which type a number is
  - To find out, use the function ```type(X)```, where ```X``` is your variable
  - Integers can usually be automatically converted to floating point numbers
  - Floating point numbers are usually **not** converted automatically
    - Computer doesn't know how to do it
<img src="computer_says_no.png" align=right width=300>
    - round the number?
    - just cut off the decimals? (truncation)
    - You have to specify:
      - ```int(x)``` truncates x (cuts off the decimal values)
      - ```np.round(x)``` rounds to the nearest integer

````python
x_fp = 10.9
x_int = int(10.9)
print(x_int)
````

````python
import numpy as np
x_fp = 10.9
x_1 = np.round(x_fp,0)
print(x_1) # note the decimal -- this is still a floating point number
print(type(x_1))
x_2 = int(np.round(x_fp,0))
print(x_2) # this is now an integer
print(type(x_2))
````

### Try a larger sample - Let's make a function for a normal sample

````python
def my_random_normal_dist(number_of_values,mean,sd):
    my_normal_values = np.full(number_of_values,np.nan)

    # fill the array with random values from a normal distribution
    for i in np.arange(0,number_of_values):
        my_normal_values[i] = np.random.normal(mean,sd)
    return(my_normal_values)
````

````python
bigger_sample = my_random_normal_dist(10000,50,10)
plt.hist(bigger_sample,bins=20)
````

#### Calculate the probability that a value is < X in a sample

````python
number_of_values = 1000
mean_of_dist = 0
sd_of_dist = 1
random_values = np.full(number_of_values,np.nan)
for i in np.arange(0,number_of_values):
    random_values[i] = np.random.normal(mean_of_dist,sd_of_dist)
my_cutoff = 1.645 
num_lt_x = np.sum(random_values < my_cutoff)
probability_lt_x = num_lt_x / number_of_values
print(probability_lt_x)
````

````python
# plot that
plt.hist (random_values,bins=20)
plt.axvline(my_cutoff,color="red")
````
